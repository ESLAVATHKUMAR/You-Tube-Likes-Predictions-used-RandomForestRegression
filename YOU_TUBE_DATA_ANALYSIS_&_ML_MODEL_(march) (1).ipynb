```python
c=widgets.Dropdown(
    options=video['Genre'].unique(),
    value='Sports',
    description='Genre:',
    disabled=False,
)
def graph(a):
  dr=video[video['Genre']==a]
  gt=dr.groupby('channelTitle')
  gt=gt.sum().sort_values(by='view_count',ascending=False).head()
  sns.barplot(y=gt.index,x=gt['view_count'])
  plt.title(a)
  plt.xlabel('view')# ======>>> to change y-axis and x-axis names
  plt.ylabel('channelName')
  plt.show()

```

```python
widgets.interact(graph,a=c)  #------>>>>> calling the finction
```
